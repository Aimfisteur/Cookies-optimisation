<h3 style="color: #4a86e8; font-size: 28px; font-family: 'Georgia', serif; text-shadow: 2px 2px 4px rgba(0,0,0,0.3); margin-top: 20px; border-left: 5px solid #4a86e8; padding-left: 10px;">Class Biscuit</h3>

```python
class Biscuit:
    def __init__(self, biscuit_type):
        self.biscuit_type = biscuit_type
        self.length, self.value, self.max_defects = self._configure_biscuit(biscuit_type)

    def _configure_biscuit(self, biscuit_type):
        if biscuit_type == 0:
            return 4, 6, {'a': 4, 'b': 2, 'c': 3}
        elif biscuit_type == 1:
            return 8, 12, {'a': 5, 'b': 4, 'c': 4}
        elif biscuit_type == 2:
            return 2, 1, {'a': 1, 'b': 2, 'c': 1}
        elif biscuit_type == 3:
            return 5, 8, {'a': 2, 'b': 3, 'c': 2}
        elif biscuit_type == 4:
            return 1, 0, {'a': 10, 'b': 10, 'c': 10} #This biscuit serves for some heuristics
        else:
            return 1, 0, {'a': 10, 'b': 10, 'c': 10}

    def biscuit_type(self):
        return self.biscuit_type
    
    def value(self):
        return self.value
    
    def length(self):
        return self.length
    
    def max_defects(self):
        return self.max_defects
    
    def __str__(self):
        return f"Biscuit Type: {self.biscuit_type}, Length: {self.length}, Value: {self.value}, Max Defects: {self.max_defects}"
```

<h3 style="color: #4a86e8; font-size: 28px; font-family: 'Georgia', serif; text-shadow: 2px 2px 4px rgba(0,0,0,0.3); margin-top: 20px; border-left: 5px solid #4a86e8; padding-left: 10px;">Class Dough</h3>

```python
class Dough:
    def __init__(self, length):
        self.LENGTH = length
        self.defects_list = []  # tuples (position, class)
        self.biscuits_placed = []  # tuples (position, biscuit_type, valid)

    def add_defect(self, position, defect_class):
        if 0 <= position < self.LENGTH:
            self.defects_list.append((position, defect_class))
        else:
            raise ValueError("Defect position out of dough range")

    def count_defects(self, position, length):
        defect_counts = {}
        for pos, cls in self.defects_list:
            if position <= pos < position + length:
                defect_counts[cls] = defect_counts.get(cls, 0) + 1
        return defect_counts

    def place_biscuits(self, biscuits):
        self.biscuits_placed.clear()
        current_position = 0
        for biscuit in biscuits:
            if current_position + biscuit.length > self.LENGTH:
                break  # Stop if there's no enough space left for the current biscuit
    
            defect_counts = self.count_defects(current_position, biscuit.length)
            valid = all(defect_counts.get(cls, 0) <= biscuit.max_defects[cls] for cls in biscuit.max_defects)
            self.biscuits_placed.append((current_position, biscuit.biscuit_type, valid))
            current_position += biscuit.length
            self.biscuits_placed = sorted(self.biscuits_placed, key=lambda x: x[0])


    def place_biscuits_GA(self, biscuit_positions):
        self.biscuits_placed.clear()
        for position, biscuit_type in biscuit_positions:
            biscuit = Biscuit(biscuit_type)
            if position + biscuit.length > self.LENGTH:
                break  
            defect_counts = self.count_defects(position, biscuit.length)
            valid = all(defect_counts.get(cls, 0) <= biscuit.max_defects[cls] for cls in biscuit.max_defects)
            self.biscuits_placed.append((position, biscuit.biscuit_type, valid))
            self.biscuits_placed = sorted(self.biscuits_placed, key=lambda x: x[0])

    
    def get_positions(self):
        for position, biscuit_type, is_valid in self.biscuits_placed:
            print(f"biscuit type: {biscuit_type}, position: {position}")
    
    def calculate_total_value(self, solution):
        total_value = 0
        for position, biscuit_type, is_valid in solution:
            defect = self.count_defects(current_position, biscuit.length)
            print(f"position: {position}, biscuit_type: {biscuit_type}, defect: {defect}, isValid: {is_valid}")
            if is_valid:
                total_value += self.biscuits[biscuit_type].value
        return total_value

    def calculate_own_value(self, biscuits):
        total_value = 0
        for position, biscuit_type, is_valid in self.biscuits_placed:
            if is_valid:
                total_value += biscuits[biscuit_type].value
        return total_value

    def validate_biscuits_placement(self, biscuits):
        last_position = -1
        for position, biscuit_type, is_valid in self.biscuits_placed:
            biscuit = biscuits[biscuit_type]

            # Checks for overlapping
            if position < last_position:
                return False, f"Overlapping detected at position {position} for biscuit type {biscuit_type}."
            last_position = position + biscuit.length

        return True, "All biscuits are correctly placed and within defect thresholds."

    def is_overlapping(self, new_position, new_length, existing_biscuits):
        for position, biscuit_type in existing_biscuits:
            biscuit_length = self.biscuits[biscuit_type].length
            if not (new_position + new_length <= position or new_position >= position + biscuit_length):
                return True 
        return False 

    def __str__(self):
        return ("Dough Defects:\n" + "\n".join(f"Position: {pos:.2f}, Class: {cls}" for pos, cls in self.defects_list) +
                "\n\nBiscuits Placed:\n" + "\n".join(f"Position: {pos}, Biscuit Type: {b_type}, Valid: {valid}" 
                                                      for pos, b_type, valid in self.biscuits_placed))

    def defects(self):
        return self.defects_list

    def biscuits(self):
        return self.biscuits_placed
```

    <h3 style="color: #2c7be5; font-size: 22px; font-family: 'Verdana', sans-serif; margin-bottom: 15px;">Dynamic programming approach  :</h3>
    <p style="line-height: 1.6; color: #333; font-family: 'Arial', sans-serif;">This problem can be solved using a dynamic programming approach in <strong>O(n) time complexity and O(n^2) space complexity</strong>. In fact we know that the goal is to use algorithms that we saw in class but as a reference it is really usefull to find the value of the best appoach as a comparison.
</p>
    <p style="line-height: 1.6; color: #333; font-family: 'Arial', sans-serif;"><strong>Here is the main idea :</strong>
</p>
    <p style="line-height: 1.6; color: #333; font-family: 'Arial', sans-serif;"><strong>The best possible value at the i-th position, \( V(i) \), is the maximum value achievable by considering all possible placements of biscuits up to the i-th position. This can be calculated as the maximum value obtained by placing a biscuit of length \( L \) at the current position, and adding it to the best value at the position \( i - L \). 

where \( L \) varies over all possible biscuit lengths.
</strong>
</p>
</div></div>

```python
#Variables
baguette  = [[] for i in range(dough.LENGTH)] #Use to simplify notation here, represents defect on their positions
for _, row in defects_df.iterrows():
    baguette[int(row['x'])] = baguette[int(row['x'])]+[row['class']]
    
biscuit_types = [Biscuit(i) for i in range(4)]# all 4 biscuits type used here

maxAtindex = [[] for i in range(500)] #use to keep track of the best path at ith position
```

```python
#Usefull functions :
def get_val(roll):#get the val from a given agencement of biscuits (sum their values)
    s=0
    for b,pos in roll:
        s+=b.value
    return s

def canBePlaced(biscuit,i,baguette = baguette): #return True if biscuit can be placed at index i on baguette 
    ca = 0
    cb = 0
    cc = 0
    for index in range(i,biscuit.length+i):
        ca+=baguette[index].count('a')
        cb+=baguette[index].count('b')
        cc+=baguette[index].count('c')
    if  ca<= biscuit.max_defects['a'] and  cb<= biscuit.max_defects['b'] and  cc<= biscuit.max_defects['c']:
        return True
    else:
        return False
```

```python
start = time.time()
for i in range(dough.LENGTH):
    maxVal = float("-inf")
    maxroll = []
    for b in biscuit_types:#iterate over all biscuits types
        if i>=b.length:
            roll = maxAtindex[i-b.length]#current best roll at position i-L
            currentVal = get_val(roll)
            
            if canBePlaced(b,i-b.length):#check if we can place it
                #print(b,i-b.length)
                currentVal+=b.value
            #print(currentVal)
            if currentVal>=maxVal:#if we got a better current value we replace it
                
                maxroll = roll+[(b,i-b.length)]
                maxVal = currentVal
    #print(maxroll)
    maxAtindex[i] = maxroll
print("Found in : ",time.time()-start,"seconds")
print("Value of Maximum : " , get_val(maxAtindex[-1]))
for b,x in maxAtindex[-1] : 
    print(b, " at position : ",x)
```

<a id="10"><h4 style="color: #6a737b; font-size: 24px; font-family: 'Courier New', monospace; margin-top: 15px; border-bottom: 1px dashed #6a737b; padding-bottom: 5px;">(i) Class Genetic Algorithm</h4></a>

<div style="background-color: #e8f4f8; border: 2px solid #b6d4e0; padding: 20px; border-radius: 10px; margin: 20px; box-shadow: 0 4px 6px rgba(0,0,0,0.1);">
    <h3 style="color: #2c7be5; font-size: 22px; font-family: 'Verdana', sans-serif; margin-bottom: 15px;">Remarks:</h3>
    <p style="line-height: 1.6; color: #333; font-family: 'Arial', sans-serif;">This is the basis of our GA reasoning, which means we will implement sub-classes from this one as a upper class. Our results will vary from the original functions emanating from this class. You can find its description in the deliverable, but there are comments resuming each function.
</p>

</div>

```python
class GeneticAlgorithm:

    #Obviously initialisation
    def __init__(self, dough, biscuits, population_size, mutation_rate, crossover_rate):
        self.dough = dough
        self.biscuits = biscuits
        self.population_size = population_size
        self.mutation_rate = mutation_rate
        self.crossover_rate = crossover_rate
        self.population = []
        self.initialize_population()

    #works with random solution to generate populations totally randomly
    def initialize_population(self):
        for _ in range(self.population_size):
            individual = self.random_solution()
            self.population.append(individual)

    #Generates a random solution, with more chances to summon biscuits 3 than 0s and 1s than 2, as our heuristic tells us
    def random_solution(self):
        biscuits = [3, 3, 3, 3] + [0, 0, 1, 1] + [2]
        solution = []
        position = 0
        while position < self.dough.LENGTH:
            biscuit_type = random.choice(biscuits)
            solution.append((position, biscuit_type))
            position += self.biscuits[biscuit_type].length
        return solution

    #Calculates the value of an individual, we did not go through the class Dough because that would make us lose Space Complexity.
    def fitness(self, individual):
        total_value = 0
        for position, biscuit_type in individual:
            biscuit = self.biscuits[biscuit_type]
            # checks if biscuit fits in the remaining dough and meets defect requirements
            if position + biscuit.length <= self.dough.LENGTH:
                defects = self.dough.count_defects(position, biscuit.length)
                if all(defects.get(cls, 0) <= biscuit.max_defects[cls] for cls in biscuit.max_defects):
                    total_value += biscuit.value
        return total_value

    #simple roulette wheel selection
    def selection(self):
        max_fitness = sum(self.fitness(ind) for ind in self.population)
        pick = random.uniform(0, max_fitness)
        current = 0
        for individual in self.population:
            current += self.fitness(individual)
            if current > pick:
                return individual

    #This is a ONE POINT Crossover, that makes sure that none of the "cuts" made to combine the parents cuts a biscuit in half and causes overlaps
    @staticmethod
    def crossover(parent1, parent2, max_attempts = 500):
        parent1 = sorted(parent1, key=lambda x: x[0])
        parent2 = sorted(parent2, key=lambda x: x[0])
        biscuits = {
            0: Biscuit(0),
            1: Biscuit(1),
            2: Biscuit(2),
            3: Biscuit(3),
            4: Biscuit(4)
        }
        #Finds the valid points
        def find_valid_crossover_points(parent, biscuits):
            valid_points = []
            for i in range(1, len(parent)):
                prev_position, prev_type = parent[i-1]
                position, _ = parent[i]
                if position - (prev_position + biscuits[prev_type].length) > 0:
                    valid_points.append(i)
            return valid_points

        #Creates the child with the found crossover point
        def create_child(p1, p2, crossover_point):
            child = [None] * len(p1)
            child[:crossover_point] = p1[:crossover_point]
            child[crossover_point:] = [b for b in p2[crossover_point:] if b not in child[:crossover_point]]
            return child

        for _ in range(max_attempts):
            valid_points_parent1 = find_valid_crossover_points(parent1, biscuits)
            valid_points_parent2 = find_valid_crossover_points(parent2, biscuits)

            if valid_points_parent1 and valid_points_parent2:
                crossover_point1 = random.choice(valid_points_parent1)
                crossover_point2 = random.choice(valid_points_parent2)

                child1 = create_child(parent1, parent2, crossover_point1)
                child2 = create_child(parent2, parent1, crossover_point2)

                return child1, child2
        #In case no valid point is found:
        return parent1, parent2

    #To keep diversity, we need to insert mutation:
    # simple mutation where a random biscuit's type or position might change
    def mutate(self, individual):
        if len(individual) > 0 and random.random() < self.mutation_rate:
            mutate_index = random.randint(0, len(individual) - 1)
            if random.random() > 0.5:
                individual[mutate_index] = (individual[mutate_index][0], random.choice(list(self.biscuits.keys())))
            else:
                position = individual[mutate_index][0]
                new_position = max(0, min(self.dough.LENGTH - self.biscuits[individual[mutate_index][1]].length, position + random.choice([-1, 1])))
                individual[mutate_index] = (new_position, individual[mutate_index][1])
        return individual
        
    #Defines how the population will evolve through generations
    def evolve(self):
        new_population = []
        while len(new_population) < self.population_size:
            parent1 = self.selection()
            parent2 = self.selection()
            if random.random() < self.crossover_rate:
                offspring1, offspring2 = self.crossover(parent1, parent2)
            else:
                offspring1, offspring2 = parent1, parent2
            offspring1 = self.mutate(offspring1)
            offspring2 = self.mutate(offspring2)
            new_population.extend([offspring1, offspring2])
        self.population = new_population[:self.population_size] 
```

```python
class GeneticElitism(GeneticAlgorithm):
    def __init__(self, dough, biscuits, population_size, mutation_rate, crossover_rate):
        super().__init__(dough, biscuits, population_size, mutation_rate, crossover_rate)

    def selection(self):
        max_fitness = sum(self.fitness(ind) for ind in self.population)
        pick = random.uniform(0, max_fitness)
        current = 0
        for individual in self.population:
            current += self.fitness(individual)
            if current > pick:
                return individual
```

<div style="background-color: #e8f4f8; border: 2px solid #b6d4e0; padding: 20px; border-radius: 10px; margin: 20px; box-shadow: 0 4px 6px rgba(0,0,0,0.1);">

    <p style="line-height: 1.6; color: #333; font-family: 'Arial', sans-serif;">Using tournament selection system, we take samples of the population ranked to sort out the results
</p>

</div>

```python
class GeneticTournament(GeneticAlgorithm):
    def __init__(self, dough, biscuits, population_size, mutation_rate, crossover_rate):
        super().__init__(dough, biscuits, population_size, mutation_rate, crossover_rate)
        self.tournament_size = 5
        
    def selection(self, elite_size):
        pop_ranked = [(individual, self.fitness(individual)) for individual in self.population]
        pop_ranked.sort(key=lambda x: x[1], reverse=True)
        selection_results = [individual[0] for individual in pop_ranked[:elite_size]]
        tournament_size = 10
        
        for _ in range(len(pop_ranked) - elite_size):
            tournament = random.sample(pop_ranked, tournament_size)
            selection_results.append(sorted(tournament, key=lambda x: x[1], reverse=True)[0][0])
        return selection_results
        
    @staticmethod
    def mutate(individual, mutation_rate):
        for swapped in range(len(individual)):
            if random.random() < mutation_rate:
                swap_with = random.randint(0, len(individual) - 1)
                individual[swapped], individual[swap_with] = individual[swap_with], individual[swapped]
        return individual

    def mutate_population(self, population):
        return [self.mutate(ind, self.mutation_rate) for ind in population]

    def evolve(self):
        elite_size = int(self.population_size * 0.2)
        selected = self.selection(elite_size)
        offspring = []
        for i in range(0, self.population_size - elite_size, 2):
            parent1 = selected[i % len(selected)]
            parent2 = selected[(i+1) % len(selected)]
            child1, child2 = GeneticAlgorithm.crossover(parent1, parent2)
            offspring.append(child1)
            if len(offspring) < self.population_size - elite_size:
                offspring.append(child2)
        self.population = selected[:elite_size] + self.mutate_population(offspring)
```

```python
def main():
    biscuits = {
        0: Biscuit(0),
        1: Biscuit(1),
        2: Biscuit(2),
        3: Biscuit(3),
        4: Biscuit(4)
    }
    dough = Dough(500)

    for _, row in defects_df.iterrows():
        dough.add_defect(position=row['x'], defect_class=row['class'])

    population_size = 150
    mutation_rate = 0.1
    crossover_rate = 0.5
    GA = GeneticTournament(dough, biscuits, population_size, mutation_rate, crossover_rate)
    num_generations = 100
    
    for generation in range(num_generations):
        GA.evolve()
        print(f"Generation {generation + 1} complete.")

    best_solution = max(GA.population, key=GA.fitness)
    total_value = GA.fitness(best_solution)
    print("Best Solution Total Value:", total_value)
    print("Best Solution Configuration:", best_solution)
    return best_solution

if __name__ == "__main__":
    best_solution = main()
```

```python
class UniformCrossoverGA(GeneticAlgorithm):
    
    def __init__(self, dough, biscuits, population_size, mutation_rate, crossover_rate):
        super().__init__(dough, biscuits, population_size, mutation_rate, crossover_rate)
        self.tournament_size = 15

    def selection(self, elite_size):
        pop_ranked = [(individual, self.fitness(individual)) for individual in self.population]
        pop_ranked.sort(key=lambda x: x[1], reverse=True)
        selection_results = [individual[0] for individual in pop_ranked[:elite_size]]
        tournament_size = 15
        
        for _ in range(len(pop_ranked) - elite_size):
            tournament = random.sample(pop_ranked, tournament_size)
            selection_results.append(sorted(tournament, key=lambda x: x[1], reverse=True)[0][0])
        return selection_results

    @staticmethod
    def uniform_crossover(parent1, parent2, max_attempts=500):
        biscuits = {
            0: Biscuit(0),
            1: Biscuit(1),
            2: Biscuit(2),
            3: Biscuit(3),
            4: Biscuit(4)
        }
        def find_valid_crossover_points(parent):
            valid_points = []
            for i in range(1, len(parent)):
                prev_position, prev_type = parent[i - 1]
                position, _ = parent[i]
                if position - (prev_position + biscuits[prev_type].length) > 0:
                    valid_points.append(i)
            return valid_points
    
        def create_child(p1, p2, crossover_point):
            child = p1[:crossover_point] + [b for b in p2[crossover_point:] if b not in p1[:crossover_point]]
            return child
    
        parent1 = sorted(parent1, key=lambda x: x[0])
        parent2 = sorted(parent2, key=lambda x: x[0])
    
        for _ in range(max_attempts):
            valid_points_parent1 = find_valid_crossover_points(parent1)
            valid_points_parent2 = find_valid_crossover_points(parent2)
    
            if valid_points_parent1 and valid_points_parent2:
                crossover_point1 = random.choice(valid_points_parent1)
                crossover_point2 = random.choice(valid_points_parent2)
    
                child1 = create_child(parent1, parent2, crossover_point1)
                child2 = create_child(parent2, parent1, crossover_point2)
    
                return child1, child2
    
        return parent1, parent2

    def evolve(self):
        
        elite_size = int(self.population_size * 0.2)
        selected = self.selection(elite_size)
        offspring = []
        
        new_population = selected[:elite_size]
    
        while len(new_population) < self.population_size:
            parent1 = selected[random.randint(0, elite_size - 1)]
            parent2 = selected[random.randint(0, elite_size - 1)]
            child1, child2 = self.uniform_crossover(parent1, parent2)
            new_population.append(child1)
            if len(new_population) < self.population_size:
                new_population.append(child2)
    
        self.population = new_population
```

```python
def main():
    biscuits = {
        0: Biscuit(0),
        1: Biscuit(1),
        2: Biscuit(2),
        3: Biscuit(3),
        4: Biscuit(4)
    }
    dough = Dough(500)

    for _, row in defects_df.iterrows():
        dough.add_defect(position=row['x'], defect_class=row['class'])

    population_size = 150
    mutation_rate = 0.1
    crossover_rate = 0.5
    GA = UniformCrossoverGA(dough, biscuits, population_size, mutation_rate, crossover_rate)
    num_generations = 100
    
    for generation in range(num_generations):
        GA.evolve()
        print(f"Generation {generation + 1} complete.")

    best_solution = max(GA.population, key=GA.fitness)
    total_value = GA.fitness(best_solution)
    print("Best Solution Configuration:", best_solution)
    
    return best_solution

if __name__ == "__main__":
    best_solution = main()
```

    <p style="line-height: 1.6; color: #333; font-family: 'Arial', sans-serif;">A greedy search will function linearly to find a solution according to an heuristic decided. Here is our main code, changing its heursitic depending on parameters, and after a quick program allowing our code to launch and execute.

```python
def greedy_biscuit_placement(dough, biscuits, heuristic = 'O'):
    
    if heuristic == 'value':
        sorted_biscuits = sorted(biscuits.values(), key=lambda b: b.value, reverse=True)
    elif heuristic == 'valength':
        sorted_biscuits = sorted(biscuits.values(), key=lambda b: b.value / b.length, reverse=True)
    else:
        sorted_biscuits = sorted(biscuits.values(), key=lambda b: b.value / (b.length*sum(b.max_defects.values())), reverse=True)
    
    total_value = 0
    current_position = 0
    placed_biscuits = []

    while current_position < dough.LENGTH:
        placed = False
        for biscuit in sorted_biscuits:
            # checks if biscuit fits in the remaining dough and meets defect requirements
            if current_position + biscuit.length <= dough.LENGTH:
                defects = dough.count_defects(current_position, biscuit.length)
                if all(defects.get(cls, 0) <= biscuit.max_defects[cls] for cls in biscuit.max_defects):
                    
                    placed_biscuits.append((current_position, biscuit.biscuit_type))
                    total_value += biscuit.value
                    current_position += biscuit.length
                    placed = True
                    break
            if not placed:
                current_position +=1

    return total_value, placed_biscuits
```

```python
def main(heuristic):
    dough = Dough(length=500)
    for _, row in defects_df.iterrows():
        dough.add_defect(position=row['x'], defect_class=row['class'])
        
    biscuits = {
        0: Biscuit(0),
        1: Biscuit(1),
        2: Biscuit(2),
        3: Biscuit(3)
    }
    
    total_value, placed_biscuits = greedy_biscuit_placement(dough, biscuits,heuristic)
    print("Total Value:", total_value)
    print("Placed Biscuits:", placed_biscuits)
```

<div style="background-color: #e8f4f8; border: 2px solid #b6d4e0; padding: 20px; border-radius: 10px; margin: 20px; box-shadow: 0 4px 6px rgba(0,0,0,0.1);">
    <p style="line-height: 1.6; color: #333; font-family: 'Arial', sans-serif;"><strong>Operations Research Tools</strong>, is an open-source software suite created by Google for solving combinatorial optimization problems. With this we can easily and fastly find good solutions or even optimum for CSP. Here is our implementation : 
</p>

</div>

<h4 style="color: #6a737b; font-size: 24px; font-family: 'Courier New', monospace; margin-top: 15px; border-bottom: 1px dashed #6a737b; padding-bottom: 5px;">(i) Import librairies</h4></a>


```python
from ortools.sat.python import cp_model
model = cp_model.CpModel()
```

<h4 style="color: #6a737b; font-size: 24px; font-family: 'Courier New', monospace; margin-top: 15px; border-bottom: 1px dashed #6a737b; padding-bottom: 5px;">(ii) Create variables</h4></a>

```python
#Variables
biscuit_types = [Biscuit(i) for i in range(4)]
for b in biscuit_types:
    print(b)

max_pos = dough.LENGTH - min(biscuit.length for biscuit in biscuit_types)#maximum position for the smallest biscuit
biscuit_vars = {}#We want to create a dictionnary that associate True if biscuit type b is placed at position p (b,p), False otherwise
for biscuit in biscuit_types:
    for start_pos in range(max_pos + 1):
        biscuit_vars[(biscuit.biscuit_type, start_pos)] = model.NewBoolVar(f'biscuit_{biscuit.biscuit_type}_start_{start_pos}')
```

```python
print(biscuit_vars[(0,3)])
```

<h4 style="color: #6a737b; font-size: 24px; font-family: 'Courier New', monospace; margin-top: 15px; border-bottom: 1px dashed #6a737b; padding-bottom: 5px;">(iii) Create Constraints</h4></a>

```python
#No overlapping biscuits
for x in range(dough.LENGTH):
    # Initialize a constraint sum for each position x
    constraint_sum = 0

    for b in biscuit_types:
        for start_pos in range(max(0, x - b.length + 1), min(x + 1, max_pos + 1)):
            if x >= start_pos and x < start_pos + b.length:
                constraint_sum += biscuit_vars[(b.biscuit_type, start_pos)]

    # Add the constraint to the model
    model.Add(constraint_sum <= 1)
```

```python
# Defects within acceptable limits

for biscuit in biscuit_types:
    for start_pos in range(max_pos + 1):
        for defect_type, max_allowed in biscuit.max_defects.items():
            # Initialize defect count for each biscuit type, start position, and defect type
            defect_count = 0

            # Count defects within the biscuit length
            for x, b in dough.defects_list:
                if b == defect_type and start_pos <= x < start_pos + biscuit.length:
                    defect_count += 1

            # Add the constraint to the model
            model.Add(defect_count * biscuit_vars[(biscuit.biscuit_type, start_pos)] <= max_allowed)
```

<h4 style="color: #6a737b; font-size: 24px; font-family: 'Courier New', monospace; margin-top: 15px; border-bottom: 1px dashed #6a737b; padding-bottom: 5px;">(iv) Objective function to maximize</h4></a>

```python
 # Maximize the total value of the biscuits
total_value = sum(biscuit_vars[(biscuit.biscuit_type, start_pos)] * biscuit.value 
                      for biscuit in biscuit_types 
                      for start_pos in range(dough.LENGTH - biscuit.length + 1))#AND Length has to be in the 500 range ! 
model.Maximize(total_value)
```

<h4 style="color: #6a737b; font-size: 24px; font-family: 'Courier New', monospace; margin-top: 15px; border-bottom: 1px dashed #6a737b; padding-bottom: 5px;">(v) Solve the model</h4></a>

```python
start =time.time()
solver = cp_model.CpSolver()
status = solver.Solve(model)
print("solved in : ", time.time()-start,"seconds")
```

```python
if status in (cp_model.OPTIMAL, cp_model.FEASIBLE):
    if status==cp_model.OPTIMAL:
        print("The solution found is optimal !")
    if status==cp_model.FEASIBLE:
        print("The solution found is feasible !")
solution = []
for biscuit in biscuit_types:
    for start_pos in range(max_pos + 1):
        if solver.Value(biscuit_vars[(biscuit.biscuit_type, start_pos)]) == 1:#if 1/True this mean that it has been placed
            solution.append((biscuit.biscuit_type, start_pos))
print("The objective function value found is : ",solver.ObjectiveValue()," and the solution found is : ",solution)
```
